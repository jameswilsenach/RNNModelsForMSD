```python
units = [100,200,300]
lays = [1,2,3]
MSD10_L_HU = []
for k in range(len(lays)):
    MSD_list = []
    for j in range(len(units)):
        MSD_temp = net.SimpleSimulation(layers=lays[k],num_hidden=units[j],num_epochs=10,out=0)
        MSD_temp.run_session()
        MSD_list.append(MSD_temp)
    MSD10_L_HU.append(MSD_list)


MSD25_L_HU = []
for k in range(len(lays)):
    MSD_list = []
    for j in range(len(units)):
        MSD_temp = net.SimpleSimulation(layers=lays[k],num_hidden=units[j],num_epochs=10,out=0,provider=1)
        MSD_temp.run_session()
        MSD_list.append(MSD_temp)
    MSD25_L_HU.append(MSD_list)
        
        
```
